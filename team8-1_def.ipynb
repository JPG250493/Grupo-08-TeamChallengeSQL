```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("data/sql-murder-mystery.db")
```

```python
def sql(query):
    return pd.read_sql_query(query, conn)
```

```python
consulta_1 = '''
SELECT *
FROM crime_scene_report
WHERE type = 'murder'
  AND date = '20180115'
  AND city = 'SQL City';
  '''
pd.read_sql_query(consulta_1, conn)
```

Security footage shows that there were 2 witnesses. The first witness lives at the last house on "Northwestern Dr". The second witness, named Annabel, lives somewhere on "Franklin Ave".

```python
consulta_2 = '''
SELECT *
FROM person as per
LEFT JOIN interview as inter
ON per.id = inter.person_id
WHERE address_street_name  = 'Northwestern Dr'
ORDER BY address_number desc
'''
pd.read_sql_query(consulta_2, conn)
```

```python
pd.read_sql_query(consulta_2, conn) ["transcript"][0]
```

```python
consulta_3 = '''
SELECT *
FROM person as per
LEFT JOIN interview as inter
ON per.id = inter.person_id
WHERE address_street_name  = 'Franklin Ave'
AND name like 'Annabel%'
'''

pd.read_sql_query(consulta_3, conn) 
```

```python
pd.read_sql_query(consulta_3, conn) ["transcript"][0]
```

```python
consulta_4 = '''
Select * 
FROM get_fit_now_member gfnm 
LEFT JOIN get_fit_now_check_in gfnci 
ON gfnm.id = gfnci.membership_id
WHERE
gfnm.membership_status = 'gold'
AND
gfnci.membership_id like '48Z%'
AND
gfnci.check_in_date = '20180109'
'''

pd.read_sql_query(consulta_4, conn)
```

```python
consulta_5 = '''
SELECT * 
FROM get_fit_now_member gfnm 
LEFT JOIN get_fit_now_check_in gfnci 
ON gfnm.id = gfnci.membership_id
LEFT JOIN person per
ON gfnm.person_id = per.id
RIGHT JOIN drivers_license dl 
ON per.license_id  = dl.id
WHERE 
gfnm.membership_status = 'gold'
AND 
gfnci.membership_id like '48Z%'
AND 
gfnci.check_in_date = '20180109'
'''

pd.read_sql_query(consulta_5, conn)
```
